```python
import csv
with open('dune.csv', 'r',encoding='utf-8') as f, open('dune_long.csv', 'w',encoding='utf-8',newline='') as f2:

    reader = csv.reader(f)
    writer = csv.writer(f2)
    for row in reader:
        text = row[0]
        if len(text) > 20:
            writer.writerow([text])
```
